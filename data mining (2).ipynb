```python
import pandas as pd
import numpy as np

# 3.1 Memuat dataset 
df = pd.read_csv(r'C:\Users\USER\OneDrive\Desktop\cay\Dataset AG - Afifah Putri Mevtilina.csv', sep=';')

# 3.2 Mengidentifikasi missing values
print("\nMissing Values per Kolom Sebelum Cleaning:")
print(df.isnull().sum())

# 3.3 Menangani Noise Data (Outliers)
# Menggunakan Z-score untuk mendeteksi outliers dan menghapusnya jika diperlukan
from scipy import stats

def remove_outliers_zscore(df, threshold=3):
    z_scores = np.abs(stats.zscore(df.select_dtypes(include='number')))
    return df[(z_scores < threshold).all(axis=1)]

df_cleaned = remove_outliers_zscore(df)

print("\nMissing Values per Kolom Setelah Cleaning:")
print(df_cleaned.isnull().sum())

print(f"\nJumlah data setelah menghapus outliers: {df_cleaned.shape[0]}")
```

```python
import pandas as pd

# Contoh DataFrame
data = {
    'category_column': ['A', 'B', 'A', 'B', 'A'],
    'numeric_column': [10, 20, 30, 40, 50]
}
df = pd.DataFrame(data)

# Melakukan agregasi data berdasarkan kategori
df_aggregated = df.groupby('category_column').agg({
    'numeric_column': ['mean', 'sum', 'count']
}).reset_index()

# Mengganti nama kolom hasil agregasi
df_aggregated.columns = ['_'.join(col).strip() if isinstance(col, tuple) else col for col in df_aggregated.columns.values]

# Menampilkan DataFrame yang telah diperbarui
print(df_aggregated)
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.preprocessing import LabelEncoder

# Load dataset
df = pd.read_csv(r'C:\Users\USER\OneDrive\Desktop\cay\Dataset AG - Afifah Putri Mevtilina.csv')

np.seterr(divide='ignore', invalid='ignore')
le = LabelEncoder()

# Misalkan X adalah fitur dan y adalah target
X = df.drop(['Longitude'], axis=1)
y = df['Longitude']

# Initialize LabelEncoder
le = LabelEncoder()

# Encode non-numeric di x
for column in X.columns:
    if X[column].dtype == 'object':  
        X[column] = le.fit_transform(X[column])
    
# Feature Selection menggunakan SelectKBest
selector = SelectKBest(score_func=f_classif, k=2
                      )
X_new = selector.fit_transform(X, y)

# Mendapatkan nama fitur yang terpilih
selected_features = X.columns[selector.get_support()]
print("Fitur terpilih:")
print(selected_features)
```
